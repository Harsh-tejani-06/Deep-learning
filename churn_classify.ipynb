```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras as tf

df=pd.read_csv("churn.csv")
df.head()
df.shape
```

```python
#drop customer_id 
df.drop('customerID',inplace=True,axis='columns')
df.head()
```

```python
# df['TotalCharges'].astype('float') this is give error beacue there is some value with space
pd.to_numeric(df.TotalCharges,errors='coerce')

df.TotalCharges.isnull()
df= df[(pd.to_numeric(df.TotalCharges,errors='coerce').isnull())== False]

df['TotalCharges']=df['TotalCharges'].astype('float64')

df.dtypes
```

```python


df.replace('No internet service','No',inplace=True)
df.replace('No phone service','No',inplace=True)
for column in df:
    print(column,":",df[column].unique())

```

```python
#changing yes_no colums 
yes_no_column= ['Partner','Dependents','MultipleLines','PhoneService','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']

for i in yes_no_column:
    df[i].replace({"Yes":1,"No":0},inplace=True)
    

for column in df:
    print(column,":",df[column].unique())
```

```python
df.gender.replace({'Female':0,'Male':1},inplace=True)

col_scaling= ['MonthlyCharges','tenure','TotalCharges']

from sklearn.preprocessing import MinMaxScaler

Scale= MinMaxScaler()

df[col_scaling]=Scale.fit_transform(df[col_scaling])

df1= pd.get_dummies(data=df,columns=['InternetService','PaymentMethod','Contract'])

# yes_no_column= ["InternetService_DSL",'InternetService_Fiber optic','InternetService_No','	PaymentMethod_Bank transfer (automatic)','PaymentMethod_Credit card (automatic)']

# for i in yes_no_column:
#     df1[i].replace({"Yes":1,"No":0},inplace=True)

bool_cols = df1.select_dtypes(include='bool').columns
df1[bool_cols] = df1[bool_cols].astype(int)

df1.dtypes

X=df1.drop('Churn',axis='columns')
y=df1['Churn']


```

```python
#train model
from sklearn.model_selection import train_test_split
from tensorflow import keras
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
model =keras.Sequential([
    keras.layers.Dense(20,input_shape=(26,),activation='relu'),
    keras.layers.Dense(20,activation='relu'),
    keras.layers.Dense(20,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')

])

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=["accuracy"] )

model.fit(x_train,y_train,epochs=50)

model.evaluate(x_test,y_test)
```

```python
#using dropuout regularization 
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
model =keras.Sequential([
    keras.layers.Dense(20,input_shape=(26,),activation='relu'),
    keras.layers.Dropout(0.3),#it drop 30% neurom
    keras.layers.Dense(20,activation='relu'),
        keras.layers.Dropout(0.4),#it drop 30% neurom
    keras.layers.Dense(20,activation='relu'),
        keras.layers.Dropout(0.5),#it drop 30% neurom
    keras.layers.Dense(1,activation='sigmoid')

])

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=["accuracy"] )

model.fit(x_train,y_train,epochs=50)

model.evaluate(x_test,y_test)
```
