```python
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow import keras
import tensorflow as tf 
# Load the dataset
df = pd.read_csv('D:\Data science\machine learning\logistic_regression.csv')

# Define features and labels
X = df[['Studied', 'Slept']]
y = df['Passed']

# Train-test split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

tb_callback =tf.keras.callbacks.TensorBoard(log_dir='logs/',histogram_freq=1)
# Build the model
# Sequential([...])	You're building a model layer-by-layer in a linear stack.
# Dense(1)	A fully connected (dense) layer with 1 neuron.
# kernel_initializer='ones' All initial weights are set to 1. Usually 'glorot_uniform' is used, but 'ones' is okay for experiments.
# bias_initializer='zeros'	Biases are initialized to 0.
#activation='sigmoid'	Outputs a value between 0 and 1 — used for binary classification.

model = keras.Sequential([
    keras.layers.Dense(1, input_shape=(2,), activation='sigmoid',
                       kernel_initializer='ones',  # use 'ones' string not 1
                       bias_initializer='zeros')   # use 'zeros' not 0
])

# Compile the model (fix 'matrices' -> 'metrics')
# optimizer='adam'	Adam optimizer automatically adjusts learning rates for better performance. Used to update weights.
model.compile(optimizer='sgd',
              loss='binary_crossentropy',
              metrics=['accuracy'])  # <-- FIXED HERE

# Train the model
model.fit(x_train, y_train, epochs=200,callbacks=[tb_callback])


```
